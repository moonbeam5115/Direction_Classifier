```python
train_batches = ImageDataGenerator().flow_from_directory(train_path, target_size=(224, 224), classes=['Facing_Forward', 'Facing_Left', 'Facing_Right'], batch_size=6)
valid_batches = ImageDataGenerator().flow_from_directory(valid_path, target_size=(224, 224), classes=['Facing_Forward', 'Facing_Left', 'Facing_Right'], batch_size=6)
test_batches = ImageDataGenerator().flow_from_directory(test_path, target_size=(224, 224), classes=['Facing_Forward', 'Facing_Left', 'Facing_Right'],       batch_size=6)
```

```python
imgs, labels = next(train_batches)
```

```python
plots(imgs, titles=labels)
```

TRANSFER LEARNING

```python
vgg16_model = tf.keras.applications.vgg16.VGG16()
```

```python
Dir_Class_Model = Sequential()
for layer in vgg16_model.layers[:-1]:
    Dir_Class_Model.add(layer)
```

```python
for layer in Dir_Class_Model.layers[:-1]:
    layer.trainable = False
```

```python
num_classes = 3
Dir_Class_Model.add(Dense(num_classes, activation='softmax'))
```

```python
Dir_Class_Model.summary()
```

```python
opt = tf.keras.optimizers.SGD(learning_rate=0.001)
```

```python
Dir_Class_Model.compile(
    optimizer=opt,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
Dir_Class_Model.fit(train_batches, validation_data=valid_batches, epochs=5, verbose=1)
```
